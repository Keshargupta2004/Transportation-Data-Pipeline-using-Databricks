```python
from pyspark.sql.functions import *
```

```python
orders_df = spark.table("transportation.bronze.orders")
```

```python
orders_clean = orders_df \
    .dropna() \
    .dropDuplicates()
```

```python
orders_clean = orders_df \
    .withColumn("year", year("order_date")) \
    .withColumn("month", month("order_date")) \
    .withColumn("day", dayofmonth("order_date"))
```

```python
display(orders_clean)
```

```python
order_clean=orders_clean.write.mode("overwrite").saveAsTable("transportation.silver.orders_clean")
```
